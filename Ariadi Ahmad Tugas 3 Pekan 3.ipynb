```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig,ax = plt.subplots(nrows=1, ncols=3, sharex=True, sharey=True, figsize=(20,8))

ax[0].plot(x, tan, label='Grafik tanx')
ax[0].set_title('Grafik tanx')
ax[1].plot(x, sin,color='red',label='Grafik sinx')
ax[1].set_title('Grafik sinx')
ax[2].plot(x, cos,color='green',label='Grafik cos')
ax[2].set_title('Grafik cosx')
fig.legend(loc='upper center')
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig,ax = plt.subplots(figsize=(12,8))
ax.plot(x,y)
ax.set_title('Grafik CosX')
int4 =ax.inset_axes([0.8,0.8,0.3,0.3])
int4.plot(x,y)
int4.plot(x,y4,color='black')
int4.set_title('Grafik Cos^4')
int3 =ax.inset_axes([0.8,-0.1,0.3,0.3])
int3.plot(x,y)
int3.plot(x,y3,color='green')
int3.set_title('Grafik Cos^3')
int5 =ax.inset_axes([-0.1,0.8,0.3,0.3])
int5.plot(x,y)
int5.plot(x,y5,color='green')
int5.set_title('Grafik Cos^5')
int2 =ax.inset_axes([-0.1,-0.1,0.3,0.3])
int2.plot(x,y)
int2.plot(x,y2,color='red')
int2.set_title('Grafik Cos^2')




```
