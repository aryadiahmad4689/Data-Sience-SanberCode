````python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
````

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

````python
fig, ax = plt.subplots(figsize=(13, 13))
palette="ch:r=-.2,d=.3_r"

#code here
d = sns.scatterplot(df['carat'],df['price'],hue=df['clarity'],size=df['depth'],palette=palette)
plt.legend(loc="upper left")

````

# Soal 4: Joint Plot
